```python
import pandas as pd
```

```python
# Load the CSV file
csv_file = 'export_alimconfiance@dgal.csv'
csv_df = pd.read_csv(csv_file, delimiter=';')
```

```python

# Load the Excel file
excel_file = 'base-pop-historiques-1876-2020.xlsx'
excel_df = pd.read_excel(excel_file, header=5)
```

```python
print("Columns in excel_df:", excel_df.columns)
print("Columns in csv_df:", csv_df.columns)
```

```python
merged_df = csv_df.merge(excel_df[['LIBGEO', 'PMUN20']], left_on='Libelle_commune', right_on='LIBGEO', how='left')
```

```python
merged_df.drop(columns=['LIBGEO'], inplace=True)
```

```python
merged_df
```

```python
merged_df.dropna(subset=['PMUN20'], inplace=True)
```

```python
merged_df['PMUN20'] = merged_df['PMUN20'].astype(int)
```

```python
merged_df
```

```python
len(merged_df)
```

```python
output_csv_file = 'result.csv'
merged_df.to_csv(output_csv_file, index=False)
```
